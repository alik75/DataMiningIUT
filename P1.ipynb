```python
import sklearn.datasets as datasets
```

```python
data=datasets.load_breast_cancer()
```

```python
# P1.1
print(data.DESCR)
```

```python
# P1.2
print(data.keys())
print(data.feature_names)
print(data.data)
```

```python
# P 1.3
import pandas as pd
df=pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
# P 1.4
df.describe()
```

```python
#  P 1.5
df["target"]=data.get("target")
df.set_index('target',inplace=True)
df.head()
```

```python
# P 1.6
print(df.index.value_counts())

print(data.target_names)
```

```python
# P 1.7
import sklearn.model_selection as model
X=data.data
Y=data.target
x_train,x_test,y_train,y_test=model.train_test_split(X,Y)
```

```python
# P 1.8
print("Data Size: " + str(df.shape))
print("X_Train Size: " + str(x_train.shape))
print("X_Test Size: " + str(x_test.shape))
print("Y_Train Size: " + str(y_train.shape))
print("Y_Test Size: " + str(y_test.shape))
```

```python
# P 1.9
from sklearn.neighbors import KNeighborsClassifier
neighb=KNeighborsClassifier(n_neighbors=6)
neighb.fit(x_train,y_train)
neighb.score(x_test,y_test)
```

```python
# P 1.10
y_pred=neighb.predict(x_test)
```

```python
# P 1.12
from sklearn.preprocessing import MinMaxScaler
MinMax=MinMaxScaler()
MinMax.fit(X)
norm_x_train=MinMax.transform(x_train)
norm_x_test=MinMax.transform(x_test)
# norm_y_train=MinMaxScaler(y_train)
# norm_y_test=MinMaxScaler(y_test)
```

```python
# P 1.13
neighb=KNeighborsClassifier(n_neighbors=6)
neighb.fit(norm_x_train,y_train)
```

```python
# P 1.14
print("test :" + str(neighb.score(norm_x_test,y_test)))
print("train :" + str(neighb.score(norm_x_train,y_train)))
```

```python
# P 1.15
train_accuracy=[]
test_accuracy=[]
for i in range(1,11):
    neighb=KNeighborsClassifier(n_neighbors=i)
    neighb.fit(norm_x_train,y_train)
    train_accuracy.insert(i,neighb.score(norm_x_train,y_train))
    test_accuracy.insert(i,neighb.score(norm_x_test,y_test))
    
print(train_accuracy)
print(test_accuracy)
```

```python
#  P 1.16
import matplotlib.pyplot as plt
plt.plot(train_accuracy,label="Train")
plt.plot(test_accuracy, label="Test")
plt.ylabel('Accuracy')
plt.xlabel('number of neighbor')
plt.legend(loc='upper right')
plt.show()
```
